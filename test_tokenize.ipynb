```python
import csv
import jieba
import jieba.posseg as pseg
jieba.set_dictionary('dict.txt.big')
jieba.load_userdict("userdict.txt")

'''
tokenize each chapter 
in order to build the userdict.txt
'''
#tokenize this chapter
index = 9

novel = []
with open('novel.csv', 'r', newline='', encoding='utf-8') as csvfile:
    rows = csv.DictReader(csvfile)
    temp = {
        'name':'',
        'content':''
    }
    i = 0
    for row in rows:
        novel.append({'name':row['name'],'content':row['content']})

#remove some useless words
stop_flags = {'x', 'c', 'u', 'd', 'p', 't', 'uj', 'm', 'f', 'r'}
words = pseg.cut(novel[index]['content'])
tmp = []
for word, flag in words:
    if flag not in stop_flags:
        tmp.append(word)
word_str = ' '.join(tmp)

print(word_str)
with open('tmp.txt', 'w', encoding='utf-8') as f:
    f.write(word_str)
```
