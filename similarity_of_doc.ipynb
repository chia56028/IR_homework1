```python
from gensim import corpora, models, similarities
from tokenize_to_json import *

'''
use gensim count similary
there are two method: tf-idf and lsi
but this homework does not use this file
'''
```

```python
class count_weight:
    def __init__(self):
        #variable
        self.novel = []
        self.tokenized = []

    def count_similarity(self):
        def use_tfidf(dictionary, corpus, test_vec):
            #build TF-IDF model
            tfidf = models.TfidfModel(corpus)
            index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=len(dictionary.keys()))
            sim = index[tfidf[test_vec]]
            #test code
            print('tfidf:', sim)
            
        def use_lsi(dictionary, corpus, test_vec):
            #build lsi model
            lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=200)
            index = similarities.SparseMatrixSimilarity(lsi[corpus], num_features=lsi.num_topics)
            sim = index[lsi[test_vec]]
            #test code
            print('lsi:', sim)
            
        #build dictionary including id and term
        dictionary = corpora.Dictionary(self.tokenized)
        #test code
#         print(dictionary.keys())
#         print(dictionary.token2id)

        #count fequency of each term
        corpus = [dictionary.doc2bow(term) for term in self.tokenized]
#         #test code
#         print(corpus)

        #count ch1
        test_vec = dictionary.doc2bow(self.tokenized[1])
#         #test code
#         print(tfidf[test_vec])
        use_tfidf(dictionary, corpus, test_vec)
        use_lsi(dictionary, corpus, test_vec)
        
        
    
    
```
